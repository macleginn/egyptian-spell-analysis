```r
files <- list.files(pattern = '\\.csv$')
```

```r
spells <- list()
for (f in files) {
    data <- read.csv(f, sep = '\t', stringsAsFactors = F)
    rownames(data) <- data[,1]
    data <- data[,-1]
    spell.name <- substr(f, 0, nchar(f)-4)
    spells[[spell.name]] <- data
}
```

```r
simple.distance <- function(df) {
    result <- matrix(0, nrow = nrow(df), ncol = nrow(df))
    rownames(result) <- rownames(df)
    colnames(result) <- rownames(df)
    for (i in 1:(nrow(df)-1))
        for (j in (i+1):nrow(df)) {
            # Count number of mismatches, divide by the length for [0,1] bounds
            distance <- sum( df[i,] != df[j,] ) / ncol(df)
            result[i,j] <- distance
            result[j,i] <- distance
        }
    as.dist(result)
}
```

```r
spell1 <- spells[[1]]
spell1.dist <- simple.distance(spell1)
spell1.umap <- umap(spell1.dist, n_neighbors = 2, min_dist = 1)
plot(spell1.umap[,2]~spell1.umap[,1], type = 'n', xlab = '', ylab = '')
text(spell1.umap[,1], spell1.umap[,2], labels = rownames(spell1))
```

```r
dimreduce.and.plot <- function(data, n_neighbors, title) {
    data.dist <- simple.distance(data)
    data.umap <- umap(data.dist, n_neighbors = n_neighbors, min_dist = 2)
    plot(data.umap[,2]~data.umap[,1], type = 'n', xlab = '', ylab = '', main = title)
    text(data.umap[,1], data.umap[,2], labels = rownames(data))
}
```

```r
# 3 neighbours
par(mfrow = c(2,3))
for (i in 1:length(spells))
    dimreduce.and.plot(spells[[i]], 3, names(spells)[i])
```

```r
# 2 neighbours
par(mfrow = c(2,3))
for (i in 1:length(spells))
    dimreduce.and.plot(spells[[i]], 2, names(spells)[i])
```
